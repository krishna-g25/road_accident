Accident Distribution by Year, Month, Day of the Week, and Hour:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Year', data=df)
plt.title('Accidents by Year')
plt.xticks(rotation=45)
plt.show()

month_map = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}

df['Month'] = df['Month'].apply(lambda x: month_map[x] if pd.notnull(x) and x in month_map else x)
sns.countplot(x='Month', data=df, order=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title('Accidents by Month')
plt.show()



sns.countplot(x='Day_of_Week', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Accidents by Day of the Week')
plt.show()
```

According to the above analysis:
the most accidents happend in the year 2021
the month which saw most number of accidents was november
the day on which most accidents occured was friday

```python
df['Junction_Detail'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Junctions with Most Accidents')
plt.xlabel('Junction Detail')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Accidents occur most frequently at locations that are not junctions, with a significant number happening within 20 meters of these spots. Conversely, the lowest number of accidents are recorded at mini-roundabouts

Below is how accidents are distributed in regions as well as the hotspots according to lattidue and longitude:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Accident_Severity', data=df, alpha=0.5)
plt.title('Accident Hotspots by Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

# Prepare data for heatmap (removing rows with NaN)
heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.dropna(subset=['Latitude', 'Longitude']).iterrows()]

HeatMap(heat_data).add_to(m)
```

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Hour'], bins=24, kde=False, color='teal')
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.xticks(range(0, 24))
plt.show()
```

according to the above picture . it is quite visible that most accidents occure between 1500 hours and 1700 hours

```python
df['Is_Weekend'] = df['Day_of_Week'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
```

```python
sns.countplot(x='Is_Weekend', data=df, palette='muted')
plt.title('Accidents on Weekdays vs Weekends')
plt.xlabel('Day Type')
plt.ylabel('Count')
plt.show()
sns.histplot(data=df, x='Hour', hue='Is_Weekend', bins=24, multiple='stack', palette='coolwarm')
plt.title('Accidents by Hour (Weekdays vs Weekends)')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()
```

the above figure shows that most number of accidents occured on weekdays rather than weekends
